```python
#1. make a new column for episode count
# FUNCTION TO EXTRACT THE EPISODES FROM THE TITLE 
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ')':
            check = False
            return data
        if check == True:
            data += i
        if i == '(':
            check = True
```

```python
# CREATING THE EPISODES COLUMN
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
# REPLACE THE STRING FROM THE EPISODES ( eps)
df['Episodes'] = df['Episodes'].str.replace(' eps', "")
```

```python
# TYPECASTING STRING TO INT VALUES TO PERFORM AGGRIGATION FUNCTIONS
df['Episodes'] = df['Episodes'].astype(int)
```

```python
#2. #make a new column for time stamp
df.loc[3]['Title']
```

```python
# EXTRACTING THE TIME AFTER THE BRACKET CLOSE OF THE EPISODES COUNT
def extraction_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]
            return data
```

```python
# MAKING A NEW COLUMN TOTAL TIME 
df['Total Time'] = df['Title'].apply(extraction_time)
```

```python
df.head()
```

```python
#EXTRACTING OF MONTHS FROM THE TOTAL TIME COLUMN
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Total Time'].apply(calculate_total_months)
```
